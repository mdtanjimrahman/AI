## Smart Home Automation System

Home automation system is a smart network that allows you to control and automate home function from a central interface often via smart-devices. It enhances convenience and energy efficiecy making daily life management easier.

# Agent Description:

1. Objective: The aim of the agent energy efficiency and convenience as well as cost-effectiveness.
2. List the sensors: Motion detectors, Door/Window sensors, Light sensors, Humidity sensors.
3. Sensors' values:
>Conditions -> {Motion, Temperature, Humidity} <br>Status -> {Day, Night}
4. List the actions:
 > Turn Lights on / Turn Lights off,<br>Turn AC on / Turn AC off, <br>Open up / Shut Windows down
5. Agent function: <br>**perception --> action**
> (Motion, Night) --> Lights on <br>
> (Motion, Day) --> Lights off<br>
> (No Motion, Night) --> Lights off<br>
> (No Motion, Day) --> Lights off<br>
> (Temp > 30, Day) --> AC on<br>
> (Temp < 30, Day) --> AC off<br>
> (Temp > 30, Night) --> AC on<br>
> (Temp < 30, Night) --> AC off<br>
> (Humidity High, Day/Night) --> Shut off windows<br>
> (Humidity Low, Day/Night) --> Open up windows



# Table driven agent

```python
# Store the agent fucntion as a table
table={('Motion','Night'):'Lights on',
       ('No Motion','Night'):'Lights off',
       ('Motion','Day'):'Lights off',
       ('No Motion','Night'):'Lights off',
       ('Temp > 30','Day'):'AC on',
       ('Temp < 30','Day'):'AC off',
       ('Temp > 30','Night'):'AC on',
       ('Temp < 30','Night'):'AC off',
       ('Humidity High','Day/Night'):'Shut off Windows',
       ('Humidity Low','Day/Night'):'Open up Windows'
       }
```

```python
percepts=[]  # to store percept history

def table_driven_agent(percept):
    print('Perception Received: '+ str(percept))
    percepts.append(percept) # updating percept history
    action = table[percept]
    return action
```

```python
import random

# Agent function
def agent_function(perception):
    motion, time, temp, humidity = perception

    actions = []

    # Motion and time-based light control
    if motion == 'Motion' and time == 'Night':
        actions.append('Lights on')
    elif (motion == 'No Motion' and time == 'Day') or (motion == 'Motion' and time == 'Day'):
        actions.append('Lights off')
    elif motion == 'No Motion' and time == 'Night':
        actions.append('Lights off')

    # Temperature-based AC control
    if temp > 30:
        actions.append('AC on')
    else:
        actions.append('AC off')

    # Humidity-based window control
    if humidity == 'Humidity High':
        actions.append('Shut off windows')
    else:
        actions.append('Open up windows')

    return actions

# Simulating the first perception randomly
motion = random.choice(['Motion', 'No Motion'])
time = random.choice(['Day', 'Night'])
temp = random.randint(20, 40)  # Random temperature between 20 and 40
humidity = random.choice(['Humidity High', 'Humidity Low'])

# To perceive environment repeatedly
while True:
    perception = (motion, time, temp, humidity)
    # The agent program is being called with the current perception
    actions = agent_function(perception)
    print('Perception:', perception)
    print('Actions Performed:', actions, '\n')

    # User control to allow next perception
    cmd = input('Type x to exit! or any key to continue.\n')
    if cmd == 'x':
        break

    # Simulating the next perception
    motion = random.choice(['Motion', 'NoMotion'])
    time = random.choice(['Day', 'Night'])
    temp = random.randint(20, 40)  # Random temperature between 20 and 40
    humidity = random.choice(['Humidity High', 'Humidity Low'])
```

# Model based agent

```python
# variables to store available information about the world state
world_state = {'Motion': 'Unknown', 'Time': 'Unknown', 'Temp': 'Unknown', 'Humidity': 'Unknown'}

def update_state(action, percept, model):
    motion, time, temp, humidity = percept

    # Update the model based on current perception
    model['Motion'] = motion
    model['Time'] = time
    model['Temp'] = temp
    model['Humidity'] = humidity

    return model

def model_based_reflex_agent(percept):
    motion, time, temp, humidity = percept
    global world_state

    actions = []

    # Motion and time-based light control
    if (world_state['Motion'] == 'No Motion' and
        ((world_state['Time'] == 'Day' and world_state['Temp'] <= 30) or
         (world_state['Time'] == 'Night' and world_state['Temp'] <= 30)) and
        world_state['Humidity'] == 'Humidity Low'):
        actions = 'Pause'
        return actions

    # Motion and time-based light control
    if motion == 'Motion' and time == 'Night':
        actions.append('Lights on')
    elif (motion == 'No Motion' and time == 'Day') or (motion == 'Motion' and time == 'Day'):
        actions.append('Lights off')
    elif motion == 'No Motion' and time == 'Night':
        actions.append('Lights off')

    # Temperature-based AC control
    if temp > 30:
        actions.append('AC on')
    else:
        actions.append('AC off')

    # Humidity-based window control
    if humidity == 'Humidity High':
        actions.append('Shut off windows')
    else:
        actions.append('Open up windows')

    # Update world state
    world_state = update_state(actions, percept, world_state)

    print('Perception: ' + str(percept))
    print('Actions Performed: ' + str(actions))
    print('World State: ' + str(world_state))

    return actions
```
